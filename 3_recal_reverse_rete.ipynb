```python
import numpy as np
import pandas as pd
import math
from Utils.GetFileDir import GetFileDir
from Utils.GetReactionNum import GetReactionNum

reactionnum = GetReactionNum()
txtname = GetFileDir(reactionnum, 'reversetxt')

fd=open(txtname)
all_data=fd.readlines()

for index,oneline in enumerate(all_data):
    oneline=oneline.split()
    A = float(oneline[5])
    N = float(oneline[6])
    Ea = float(oneline[7])
    EaR = Ea/1.987
    Tlow = int(oneline[3])
    Thigh = int(oneline[4])
    K0 = 0
    if Thigh != 0:
        if((Tlow <= 298) and (Thigh >= 298)):
            print(A)
            print(Ea)
            K0 = A*math.exp((-Ea)/(298*1.987))
            print(K0)
    else:
        if(Tlow == 298):
            K0 = A*math.exp((-Ea)/(298*1.987))
    splitdata = all_data[index].split()

    all_data[index] = "{:<20s}".format(splitdata[0]) + \
    "{:<30s}".format(splitdata[1])+\
    "{:<5s}".format(splitdata[2])+\
    "{:<10s}".format(splitdata[3])+\
    "{:<10s}".format(splitdata[4])+\
    "{:<10s}".format(splitdata[5])+\
    "    "+\
    "{:<10s}".format(splitdata[6])+\
    "{:<20s}".format(str(round(EaR,4)))+\
    "{:.5e}".format(K0)+\
    "    "+\
    "{:<5s}".format(splitdata[9])+\
    "{:<5s}".format(splitdata[10])+\
    "{:<5s}".format(splitdata[11])+\
    "{:<5s}".format(splitdata[12])+\
    "{:<5s}".format(splitdata[13])+\
    "{:<5s}".format(splitdata[14])+\
    "{:<5s}".format(splitdata[15])+\
    "{:<5s}".format(splitdata[16])+\
    "{:<5s}".format(splitdata[17])+\
    "{:<5s}".format(splitdata[18])+\
    "{:<5s}".format(splitdata[19])+'\n'

fd.close()

fd=open(txtname,'w')
fd.writelines(all_data)
fd.close()
```
